```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return (x+3)**2

def gradient(x):
    return 2*(x+3)

start = x = 2
rate = 0.1
iterations = 50
x_values = [x]
y_values = [function(x)]

for i in range(iterations):
    grad = gradient(x)
    x = x - rate*grad
    x_values.append(x)
    y_values.append(function(x))

x_range = np.linspace(-6,6,100)
y_range = function(x_range)

plt.figure(figsize = (10,6))
plt.plot(x_range, y_range, label = 'y = (x + 3)^2', color = 'blue')
plt.scatter(x_values, y_values, label = 'Gradient Descent Path', color = 'red')
plt.scatter(start, function(start), label = 'Starting Point (x = 2)', color = 'green')
plt.scatter(x_values[-1], y_values[-1], label = 'Local Minima', color = 'purple')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Gradient Descent to Find Local Minima of y = (x + 3)^2")
plt.show()
```
